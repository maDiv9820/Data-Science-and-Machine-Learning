```python
import csv
import numpy as np
with open('terrorismData.csv') as file:
    data = list(csv.DictReader(file,skipinitialspace = True))
killedValues = []
for info in data:
    if info['Country'] == 'United States' and info['Killed'] != '':
        killedValues.append(float(info['Killed']))
killedValues = np.array(killedValues,int)
print(np.sum(killedValues))
```
