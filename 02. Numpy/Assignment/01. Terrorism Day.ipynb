```python
import csv
import numpy as np
with open('/home/madiv/Data Science and Machine Learning/02. Numpy/terrorismData.csv') as file:
    data = list(csv.DictReader(file,skipinitialspace = True))
days,attacks = [],[]
for info in data:
    days.append(int(info['Day']))
    attacks.append(1)
days = np.array(days)
attacks = np.array(attacks)
attack = attacks[days >= 10 & days <= 20]
print(np.sum(attack))
```
