```python
def sortAsPerDifferentColumn(array):
    return array[1]
import csv
import numpy as np
with open('/home/madiv/Data Science and Machine Learning/02. Numpy/terrorismData.csv') as file:
    data = list(csv.DictReader(file,skipinitialspace = True))
peopleKilled,peopleWounded,cities = [],[],[]
for info in data:
    country = info['Country']
    killed = float(info['Killed']) if info["Killed"] != '' else 0.0
    wounded = float(info['Wounded']) if info["Wounded"] != '' else 0.0
    city = info['City']
    if country == 'India':
        peopleKilled.append(killed)
        peopleWounded.append(wounded)
        cities.append(city)
peopleKilled = np.array(peopleKilled, dtype = int)
peopleWounded = np.array(peopleWounded, dtype = int)
casualties = peopleKilled+peopleWounded
cityWiseCasualty = {}
cityAndCasualty = []
for index in range(len(casualties)):
    city = cities[index]
    casualty = casualties[index]
    cityWiseCasualty[city] = cityWiseCasualty.get(city,0)+casualty
for city in cityWiseCasualty:
    cityAndCasualty.append([city,cityWiseCasualty[city]])
cityAndCasualty.sort(key = sortAsPerDifferentColumn,reverse = True)
print(cityAndCasualty[:5])
```
