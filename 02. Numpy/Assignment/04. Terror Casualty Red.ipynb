```python
import csv
import numpy as np
with open('/home/madiv/Data Science and Machine Learning/02. Numpy/terrorismData.csv') as file:
    data = list(csv.DictReader(file,skipinitialspace = True))
peopleKilled, peopleWounded, states = [],[],[]
for info in data:
    killed = float(info['Killed']) if info['Killed'] != '' else 0.0
    wounded = float(info['Wounded']) if info['Wounded'] != '' else 0.0
    state = info['State']
    peopleKilled.append(killed)
    peopleWounded.append(wounded)
    states.append(state)
peopleKilled = np.array(peopleKilled, dtype = int)
peopleWounded = np.array(peopleWounded, dtype = int)
states = np.array
casuality = peopleKilled+peopleWounded
casualityInRedStates = casuality[(states == 'Jharkhand')|(states == 'Odisha')|(states == 'Andhra Pradesh')|
                                 (states == 'Chhattisgarh')]
print(np.sum(casualityInRedStates))
```
