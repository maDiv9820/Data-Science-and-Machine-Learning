```python
import csv
import numpy as np
with open('/home/madiv/Data Science and Machine Learning/02. Numpy/terrorismData.csv') as file:
    data = list(csv.DictReader(file,skipinitialspace = True))
days = []
for info in data:
    days.append(float(info['Day']))
days = np.array(days,dtype = int)
frequencyAndDay = [0 for index in range(32)]
for day in np.unique(days):
    frequencyAndDay[day] = np.sum([days == day])
frequencyAndDay = np.array(frequencyAndDay)
frequentDay = np.argmax(frequencyAndDay)
print(frequentDay,frequencyAndDay[frequentDay])
```
