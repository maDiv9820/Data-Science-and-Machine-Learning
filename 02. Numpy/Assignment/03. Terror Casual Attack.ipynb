```python
import csv
import numpy as np
with open('/home/madiv/Data Science and Machine Learning/02. Numpy/terrorismData.csv') as file:
    data = list(csv.DictReader(file,skipinitialspace = True))
killedDayWise = []; woundedDayWise = []
cities = []; groups = []
for info in data:
    state = info['State']; year = int(info['Year']); month = int(info['Month'])
    killed = info['Killed']; wounded = info['Wounded']; city = info['City']; group = info['Group']
    if state == 'Jammu and Kashmir' and year == 1999 and (month >= 5 and month <= 7):
        if killed == '':
            killed = 0
        if wounded == '':
            wounded = 0
        killedDayWise.append(int(float(killed)))
        woundedDayWise.append(int(float(wounded)))
        cities.append(city)
        groups.append(group)
killedDayWise = np.array(killedDayWise, dtype = int)
woundedDayWise = np.array(woundedDayWise, dtype = int)
casualtyDayWise = killedDayWise+woundedDayWise
index = np.argmax(casualtyDayWise)
print(casualtyDayWise[index],cities[index],groups[index])
```
