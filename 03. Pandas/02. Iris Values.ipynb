```python
import numpy as np
import pandas as pd
dataset = pd.read_csv('https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv')
df = dataset.copy()
flower_types = np.unique(df['variety'])
for flower in flower_types:
    describe = df[df['variety'] == flower].describe()
    print(describe.index)
    values = ['min','max','mean']
    for value in values:
        for prop in describe.columns:
            ans = describe[str(prop)][value]
            print('{:.2f}'.format(ans),end = ' ')
        print(flower)
```
