```python
import pandas as pd
dataset = pd.read_csv('terrorismData.csv')
df = dataset[dataset['State'] == 'Jammu and Kashmir']
print(df)
city_count = df.City.value_counts()
city = city_count.index[0]
count = city_count.values[0]
groups = df.group[(df['City'] == city)&(df['Group']!='Unknown')].value_counts()
group = groups.index[0]
print(city,count,group)
```
