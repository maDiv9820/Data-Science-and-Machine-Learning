```python
def findNoOfJobsForCountry(data, Country):
    totalJobs = 0
    for values in data:
        location = values['location']
        country = location.split(',')[0].strip()
        if country == Country:
            totalJobs += 1
    return totalJobs
```

```python
import csv
with open('amazon_jobs_dataset.csv') as file:
    data = list(csv.DictReader(file, skipinitialspace = True))
jobsInCanada = findNoOfJobsForCountry(data,'CA')
print(jobsInCanada)
```
