```python
import csv
with open('amazon_jobs_dataset.csv') as file:
    data = list(csv.DictReader(file, skipinitialspace = True))
countryJobCount = {}
for values in data:
    country = values['location'].split(',')[0].strip()
    basic_qualification = values['BASIC QUALIFICATIONS']
    if country != '' and 'Java' in basic_qualification:
        countryJobCount[country] = countryJobCount.get(country,0)+1
countryWithMaxJob,maxJobs = '',0
for country in countryJobCount:
    if maxJobs < countryJobCount[country]:
        maxJobs = countryJobCount[country]
        countryWithMaxJob = country
print(countryWithMaxJob,maxJobs)
```
