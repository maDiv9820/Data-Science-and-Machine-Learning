```python
import csv
with open('amazon_jobs_dataset.csv') as file:
    data = list(csv.DictReader(file, skipinitialspace = True))
noOfJobsAsPerLanguage = {'C++': 0, 'Java': 0, 'Python': 0}
for values in data:
    basic_qualifications = values['BASIC QUALIFICATIONS']
    country = values['location'].split(',')[0].strip()
    if country == 'IN' and ('Bachelor' in basic_qualifications or 'BS' in basic_qualifications 
                           or 'BA' in basic_qualifications):
        if 'C++' in basic_qualifications:
            noOfJobsAsPerLanguage['C++'] += 1
        if 'Java' in basic_qualifications:
            noOfJobsAsPerLanguage['Java'] += 1
        if 'Python' in basic_qualifications:
            noOfJobsAsPerLanguage['Python'] += 1
maxJob,languageWithMaxJob = 0,''
for language in noOfJobsAsPerLanguage:
    if maxJob < noOfJobsAsPerLanguage[language]:
        maxJob = noOfJobsAsPerLanguage[language]
        languageWithMaxJob = language
print(language,maxJob)
```
