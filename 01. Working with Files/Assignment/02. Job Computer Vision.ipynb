```python
def findNoOfJobsForTitle(data, jobTitle):
    totalJobs = 0
    for values in data:
        if jobTitle in values['Title']:
            totalJobs += 1
    return totalJobs
```

```python
import csv
with open('amazon_jobs_dataset.csv') as file:
    data = list(csv.DictReader(file, skipinitialspace = True))
jobWithComputerVision = findNoOfJobsForTitle(data,'Computer Vision')
print(jobWithComputerVision)
```
