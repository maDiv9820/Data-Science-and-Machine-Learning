```python
def findMonthWithMaxJobs(data, year):
    maxMonth,maxJobs = '',0
    monthWiseJobs = {}
    for values in data:
        postingDate = values['Posting_date']
        Month = postingDate.split()[0].strip()
        Year = int(postingDate.split()[-1].strip())
        if Year == year:
            monthWiseJobs[Month] = monthWiseJobs.get(Month,0)+1
    for month in monthWiseJobs:
        jobs = monthWiseJobs[month]
        if jobs > maxJobs:
            maxJobs = jobs
            maxMonth = month
    return maxMonth,maxJobs
```

```python
import csv
with open('amazon_jobs_dataset.csv') as file:
    data = list(csv.DictReader(file, skipinitialspace = True))
jobsIn2018 = findMonthWithMaxJobs(data,2018)
print(jobsIn2018)
```
