```python
import csv
with open('amazon_jobs_dataset.csv') as file:
    data = list(csv.DictReader(file,skipinitialspace = True))
noOfJobs = 0
for values in data:
    basic_qualifications = values['BASIC QUALIFICATIONS']
    if 'Bachelor' in basic_qualifications or 'BS' in basic_qualifications or 'BA' in basic_qualifications:
        noOfJobs += 1
print(noOfJobs)
```
