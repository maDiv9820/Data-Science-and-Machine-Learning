```python
def findNoOfJobs(data, city):
    totalJobs = 0
    for values in data:
        location = values['location'].split(',')[-1].strip()
        if location == city:
            totalJobs += 1
    return totalJobs
```

```python
import csv
with open('amazon_jobs_dataset.csv') as file:
    data = list(csv.DictReader(file, skipinitialspace = True))
totalJobsInBanglore = findNoOfJobs(data,'Bangalore')
totalJobsInSeattle = findNoOfJobs(data,'Seattle')
print(totalJobsInBanglore,totalJobsInSeattle)
```
