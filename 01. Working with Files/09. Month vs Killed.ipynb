```python
import csv
with open('year2017.csv') as file:
    data = list(csv.DictReader(file,skipinitialspace = True))
    monthlyKilled = {}
    for values in data:
        month = values['Month']
        killed = values['Killed']
        if killed == '':
            killed = '0.0'
        killed = int(float(killed))
        monthlyKilled[month] = monthlyKilled.get(month,0)+killed
    for month in monthlyKilled:
        print(month,monthlyKilled[month])
```
