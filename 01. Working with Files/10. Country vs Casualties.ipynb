```python
import csv
with open('year2017.csv') as file:
    data = list(csv.DictReader(file,skipinitialspace = True))
    countryWiseCasualty = {}
    for values in data:
        country = values['Country']
        casualty = values['casualities']
        if casualty == '':
            casualty = '0.0'
        casualty = int(float(casualty))
        countryWiseCasualty[country] = countryWiseCasualty.get(country,0)+casualty
for country in countryWiseCasualty:
    print(country,countryWiseCasualty[country])
```
